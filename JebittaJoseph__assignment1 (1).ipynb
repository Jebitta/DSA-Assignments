1.Histogram

```python
plt.figure(figsize=(8,10))

plt.subplot(3,2,1)
plt.hist(iris['Classification'])
plt.title('Histogram of classification')
plt.xlabel('Classifications')
plt.ylabel('Frequency')

plt.subplot(3,2,2)
plt.hist(iris['SL'])
plt.title('Histogram of SepalLength')
plt.xlabel('SepalLength')
plt.ylabel('Frequency')

plt.subplot(3,2,3)
plt.hist(iris['SW'])
plt.title('Histogram of SepalWidth')
plt.xlabel('SepalWidth')
plt.ylabel('Frequency')

plt.subplot(3,2,4)
plt.hist(iris['PL'])
plt.title('Histogram of PetalLength')
plt.xlabel('PetalLength')
plt.ylabel('Frequency')

plt.subplot(3,2,5)
plt.hist(iris['PW'])
plt.title('Histogram of PetalWidth')
plt.xlabel('PetalWidth')
plt.ylabel('Frequency')

plt.tight_layout()
```

2 Bo plot

```python
plt.figure(figsize=(8,10))

plt.subplot(2,2,1)
sns.boxplot(x='Classification',y='SL',data=iris)

plt.subplot(2,2,2)
sns.boxplot(x='Classification',y='SW',data=iris)

plt.subplot(2,2,3)
sns.boxplot(x='Classification',y='PL',data=iris)

plt.subplot(2,2,4)
sns.boxplot(x='Classification',y='PW',data=iris)

plt.tight_layout()
```

3.Scatter plot

```python
colors=['red','orange','blue']
species=['Iris-setosa','Iris-virginica','Iris-versicolor']

plt.figure(figsize=(8,10))

plt.subplot(2,2,1)
for i in range(3):
  x=iris[iris['Classification']==species[i]]
  plt.scatter(x['SL'],x['SW'],c=colors[i],label=species[i])
  plt.xlabel('SL')
  plt.ylabel('SW')
  plt.legend()

plt.subplot(2,2,2)
for i in range(3):
  x=iris[iris['Classification']==species[i]]
  plt.scatter(x['PL'],x['PW'],c=colors[i],label=species[i])
  plt.xlabel('PL')
  plt.ylabel('PW')
  plt.legend()

plt.subplot(2,2,3)
for i in range(3):
  x=iris[iris['Classification']==species[i]]
  plt.scatter(x['SL'],x['PL'],c=colors[i],label=species[i])
  plt.xlabel('SL')
  plt.ylabel('PL')
  plt.legend()

plt.subplot(2,2,4)
for i in range(3):
  x=iris[iris['Classification']==species[i]]
  plt.scatter(x['SW'],x['PW'],c=colors[i],label=species[i])
  plt.xlabel('SW')
  plt.ylabel('PW')
  plt.legend()

plt.tight_layout()

              
```
